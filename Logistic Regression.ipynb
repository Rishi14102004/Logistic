```python
#Model Building:
```

```python
Y=df["Outcome"]
```

```python
# standardization
X=df[df.columns[[0,1,2,3,4,5,6,7]]]
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()
SS_X = SS.fit_transform(X)
SS_X = pd.DataFrame(SS_X)
SS_X.columns = list(X)
SS_X
```

```python
#Data Partitionoing
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(SS_X,Y,test_size=0.4,random_state=50000)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)   

Y_pred_train=model.predict(X_train)
Y_pred_test=model.predict(X_test)

from sklearn.metrics import accuracy_score
Trainingscore=accuracy_score(Y_train,Y_pred_train)
print("Training accuracy Score:",np.round(Trainingscore,2))

Testscore=accuracy_score(Y_test,Y_pred_test)
print("Test accuracy score:",np.round(Testscore,2))
```

```python
#Model Evaluation:
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(Y_test,Y_pred_test)
cm
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,recall_score,precision_score,f1_score
print("Recall score:",np.round(recall_score(Y_test,Y_pred_test),2))
print("Specificity:",np.round(recall_score(Y_pred_test,Y_test),2))
print("Precision Score:",np.round(precision_score(Y_test,Y_pred_test),2))
print("F1 Score:",np.round(f1_score(Y_test,Y_pred_test),2))
```

```python
df[["Y_prob"]]=model.predict_proba(SS_X)[ : ,1:]
df.head()
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
tpr,fpr,dummy =roc_curve(Y,df[["Y_prob"]])
```

```python
roc_auc_score(Y,df[["Y_prob"]])
```

```python
rocscore=roc_auc_score(Y,df[["Y_prob"]])
rocscore
print("Area under curve score:",np.round(rocscore,2))

plt.scatter(tpr,fpr)
plt.plot(tpr,fpr,color='red')
plt.xlabel("True Positive Rate")
plt.ylabel("False Positive Rate")
plt.show()
```

```python
from sklearn.metrics import log_loss
loss=log_loss(Y,df[["Y_prob"]])
loss
print("log loss:",np.round(loss,2))
```

```python
"""Interpretation of the Logistic Regression Coefficients
Logistic regression models the probability of survival (target variable) using the log-odds function.
Each coefficient represents the change in the log-odds of survival for a one-unit increase in the predictor, keeping all other variables constant.
Positive coefficient (Î² > 0):
    Indicates that as the predictor increases, the probability of survival increases.
Negative coefficient (Î² < 0):
    Indicates that as the predictor increases, the probability of survival decreases.
Magnitude of coefficient:
    Larger absolute values indicate a stronger influence on survival.
Odds Ratio (eáµ):
    coefficients into an interpretable form:
    Odds Ratio > 1 â†’ increases survival chances
    Odds Ratio < 1 â†’ decreases survival chances
The intercept represents the log-odds of survival when all predictors are zero and is generally not interpreted directly."""
```

```python
'''Based on the logistic regression model, the most important features influencing survival are explained below:

Gender:
A strong predictor of survival. Being female significantly increases the probability of survival compared to males. This suggests women were more likely to survive.

Passenger Class (Pclass):
Higher passenger class (e.g., first class) has a positive effect on survival. Lower classes show reduced survival probability, indicating socio-economic status played a role.

Age:
Age generally has a negative coefficient, meaning survival probability decreases as age increases. Younger passengers had higher survival chances.

Fare:
Higher fare values increase the probability of survival, often associated with higher passenger class and better access to safety measures.

Family-related variables (SibSp, Parch):
These variables show moderate influence. Traveling with a small family increases survival probability, while large family sizes may reduce it.

Less significant features:
Variables with coefficients close to zero have minimal impact and contribute little to predicting survival.'''
```
